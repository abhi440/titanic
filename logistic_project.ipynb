```python
#there are string value so we convert these variables into dummies
sex=pd.get_dummies(df['Sex'],drop_first=True)
```

```python
sex.head(5)
```

```python
embark=pd.get_dummies(df['Embarked'],drop_first=True)
```

```python
passe=pd.get_dummies(df['Pclass'],drop_first=True)
```

```python
df.head()
```

```python
df=pd.concat([df,sex,embark,passe],axis=1)
```

```python
df.head(5)
```

```python
df.drop(['PassengerId','Name','Sex','Ticket','Embarked'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.drop(['Pclass'],axis=1,inplace=True)
```

```python
#train_test_data
```

```python
x=df[['Age','SibSp','Parch','male','Q','S']]
y=df['Survived']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=1,test_size=.3)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel=LogisticRegression()
```

```python
logmodel.fit(x_train,y_train)
```

```python
predictions=logmodel.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
classification_report(y_test,predictions)
```

```python
confusion_matrix(y_test,predictions)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,predictions)
```
